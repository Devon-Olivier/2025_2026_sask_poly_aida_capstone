### Yard-Meter inter conversion

$$
\begin{align}
1~\mathrm{yd}
&= 36~\mathrm{in} \\
&= 36 \times 2.54~\mathrm{cm} \\
&= 36 \times 2.54 \times 10^{-2}~\mathrm{m} \\
&= 0.9144~\mathrm{m}
\end{align}
$$

$$
\begin{align}
1~\mathrm{m}
&= \frac{1}{36 \times 2.54 \times 10^{-2}}~\mathrm{yd} \\
&\approx 1.093613298337708~\mathrm{yd}
\end{align}
$$
